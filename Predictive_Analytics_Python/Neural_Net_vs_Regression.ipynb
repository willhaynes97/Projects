## Data Imputation Strategy

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import ExtraTreesRegressor

dataset = pd.read_csv('q3_dataset.csv')
print("Column-wise missing value percentage")
print(dataset.isnull().sum()/len(dataset)*100)

X = dataset.drop(['ViolentCrimesPerPop'],axis=1)
Y = dataset['ViolentCrimesPerPop']

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.30, random_state=11)
```

### Answer

```python
# fill all rows with 0
X_train_fill_all_0 = X_train.fillna(0)
X_test_fill_all_0 = X_test.fillna(0)

# linear regression
reg = linear_model.LinearRegression()
reg.fit(X_train_fill_all_0,Y_train)
Y_pred_test_fill_all_0 = reg.predict(X_test_fill_all_0)
# Test mean abolute error
print("Test mean absolute error filling all NaNs with 0: %.4f"
      % mean_absolute_error(Y_test, Y_pred_test_fill_all_0))

# fill all rows with column mean
X_train_fill_mean = X_train.fillna(X_train.mean())
X_test_fill_mean = X_test.fillna(X_test.mean())

# linear regression
reg = linear_model.LinearRegression()
reg.fit(X_train_fill_mean,Y_train)
Y_pred_test_fill_mean = reg.predict(X_test_fill_mean)
# Test mean abolute error
print("Test mean absolute error filling all NaNs with the column mean: %.4f"
      % mean_absolute_error(Y_test, Y_pred_test_fill_mean))

# fill all rows with column median
X_train_fill_median = X_train.fillna(X_train.median())
X_test_fill_median = X_test.fillna(X_test.median())

# linear regression
reg = linear_model.LinearRegression()
reg.fit(X_train_fill_median,Y_train)
Y_pred_test_fill_median = reg.predict(X_test_fill_median)
# Test mean abolute error
print("Test mean absolute error filling all NaNs with the column median: %.4f"
      % mean_absolute_error(Y_test, Y_pred_test_fill_median))

# fill all rows with column mode
X_train_fill_mode = X_train.fillna(X_train.mode().iloc[0])
X_test_fill_mode = X_test.fillna(X_train.mode().iloc[0])
# linear regression
reg = linear_model.LinearRegression()
reg.fit(X_train_fill_mode,Y_train)
Y_pred_test_fill_mode = reg.predict(X_test_fill_mode)
# Test mean abolute error
print("Test mean absolute error filling all NaNs with the column mode: %.4f"
      % mean_absolute_error(Y_test, Y_pred_test_fill_mode))

# Random Forest fill missing values
imp_2 = IterativeImputer(random_state=0, estimator=ExtraTreesRegressor(n_estimators=5, random_state=0))
imp_2_x_train = pd.DataFrame(imp_2.fit_transform(X_train))
imp_2_x_test = pd.DataFrame(imp_2.fit_transform(X_test))

# linear regression
reg = linear_model.LinearRegression()
reg.fit(imp_2_x_train,Y_train)
Y_pred_test_rf = reg.predict(imp_2_x_test)
# Test mean abolute error
print("Test mean absolute error filling all NaNs with the Random Forest (n_trees=5): %.4f"
      % mean_absolute_error(Y_test, Y_pred_test_rf))

# KNN to fill missing values
imp_1 = IterativeImputer(random_state=0, estimator=KNeighborsRegressor(n_neighbors=10))
imp_1_x_train = pd.DataFrame(imp_1.fit_transform(X_train))
imp_1_x_test = pd.DataFrame(imp_1.fit_transform(X_test))

# linear regression
reg = linear_model.LinearRegression()
reg.fit(imp_1_x_train,Y_train)
Y_pred_test_knn = reg.predict(imp_1_x_test)
# Test mean abolute error
print("Test mean absolute error filling all NaNs with the KNN (k=10): %.4f"
      % mean_absolute_error(Y_test, Y_pred_test_knn))

# get rid of police stats
X_train_no_polic = X_train.drop(['PolicBudgPerPop', 'PolicCars'], axis=1)
X_test_no_polic = X_test.drop(['PolicBudgPerPop', 'PolicCars'], axis=1)

# KNN to fill missing values without police
imp_1_no_polic = IterativeImputer(random_state=0, estimator=KNeighborsRegressor(n_neighbors=10))
imp_1_x_train_no_polic = pd.DataFrame(imp_1_no_polic.fit_transform(X_train_no_polic))
imp_1_x_test_no_polic = pd.DataFrame(imp_1_no_polic.fit_transform(X_test_no_polic))

# linear regression
reg = linear_model.LinearRegression()
reg.fit(imp_1_x_train_no_polic,Y_train)
Y_pred_test_knn_no_polic = reg.predict(imp_1_x_test_no_polic)
# Test mean abolute error
print("Test mean absolute error No police data and filling all NaNs with the KNN (k=10): %.4f"
      % mean_absolute_error(Y_test, Y_pred_test_knn_no_polic))
```

Answer:  
The best performing strategy was using k-nearest neighbors to predict the missing values based on the related columns. This is a better prediction of missing values then column-wise statical metrics like mean, median, or mode because a point value is best approximated by the values of the points that are closest to it. With mean and median you are factoring points that are not similar to it to predict the missing value. Both police features are missing a large amount of data, so I tried removing these columns from the dataset entirely and running KNN again, but this resulted in a worse Test mean absolute error. Therefore the police data is worth keeping around and using KNN to predict the values of the police data improves the prediction power of the model. This also hints to me that the data is missing systematically and not at random because if it was at random doing a KNN to predict the missing police values would have had about the same test mean absolute error as removing the missing police values.
