```python
def truncate_dates(sector_dict):
    tickers = list(sector_dict.values())
    return(tiingo.get_dataframe(tickers, start_date, metric_name='adjClose').dropna(axis=1))
```

```python
def returns(prices):
    rf = rf_prices
    daily_returns = prices.pct_change().dropna()
    daily_returns_rf = rf.pct_change().dropna()
    annual_returns = prices.resample('A').last().pct_change().dropna()
    annual_returns_rf = rf.resample('A').last().pct_change().dropna()
    monthly_returns = prices.resample('M').last().pct_change().dropna()
    monthly_returns_rf = rf.resample('M').last().pct_change().dropna()      
    
    daily_risk_premium = daily_returns.sub(daily_returns_rf.iloc[:,0], axis = 0)
    annual_risk_premium = annual_returns.sub(annual_returns_rf.iloc[:,0], axis = 0)
    monthly_risk_premium = monthly_returns.sub(monthly_returns_rf.iloc[:,0], axis = 0)
    
    return ([daily_risk_premium,annual_risk_premium,monthly_risk_premium])
```

```python
def ew_index_returns(prices, col_name):
    rf = rf_prices
    daily_returns = prices.pct_change().dropna()
    daily_returns_rf = rf.pct_change().dropna()
    annual_returns = prices.resample('A').last().pct_change().dropna()
    annual_returns_rf = rf.resample('A').last().pct_change().dropna()    
    monthly_returns = prices.resample('M').last().pct_change().dropna()
    monthly_returns_rf = rf.resample('M').last().pct_change().dropna()    
    
    tickers = list(prices.columns)
    w_equal = pd.Series(1/len(tickers), index=tickers)    # len(tickers) = number of risky assets                       
    equal_weight_daily_return = daily_returns.dot(w_equal)   
    equal_weight_annual_return = annual_returns.dot(w_equal)
    equal_weight_monthly_return = monthly_returns.dot(w_equal)
    
    daily_risk_premium = equal_weight_daily_return.sub(daily_returns_rf.iloc[:,0], axis = 0)
    annual_risk_premium = equal_weight_annual_return.sub(annual_returns_rf.iloc[:,0], axis = 0)
    monthly_risk_premium = equal_weight_monthly_return.sub(monthly_returns_rf.iloc[:,0], axis = 0)
    
    daily_risk_premium = pd.DataFrame(daily_risk_premium, columns = [col_name])
    annual_risk_premium = pd.DataFrame(annual_risk_premium, columns = [col_name])
    monthly_risk_premium = pd.DataFrame(monthly_risk_premium, columns = [col_name])
    
    
    return ([daily_risk_premium,annual_risk_premium,monthly_risk_premium])
```

```python
rp      = rx_annual.mean()  # Select columns in the list "tickers"
vol     = rx_daily.std() * 252**0.5 
cov     = rx_daily.cov() * 252
cov_inv = np.linalg.inv(cov)

risky_assets = risk_return = pd.DataFrame()                        
risky_assets['RP']     = rp
risky_assets['Vol']    = vol      
risky_assets['Sharpe'] = risky_assets.RP / risky_assets.Vol
risky_assets['Color']  = 'orange'
risky_assets['Size']   = 100

riskfree_asset = pd.DataFrame(index=['SHY'])                        
riskfree_asset['RP']     = [0]
riskfree_asset['Vol']    = [0]       
riskfree_asset['Color']  = 'green'
riskfree_asset['Size']   = 100

w_equal = pd.Series(1/len(list(rx_annual.columns)), index=list(rx_annual.columns))    # len(tickers) = number of risky assets
equal_weight = pd.DataFrame(index=['E_W'])                        
equal_weight['RP']     = w_equal.dot(rp)
equal_weight['Vol']    = w_equal.dot(cov).dot(w_equal) ** 0.5      
equal_weight['Sharpe'] = equal_weight.RP / equal_weight.Vol
equal_weight['Color']  = 'blue'
equal_weight['Size']   = 50

n_assets     = len(risky_assets.index)
n_portfolios = 5000000

random_numbers = [ np.random.uniform(0,1,n_assets) for i in range(0,n_portfolios) ]
weights        = [ w/w.sum() for w in random_numbers ]

random = pd.DataFrame()                                      
random['RP']     = [ w.dot(rp)              for w in weights ] 
random['Vol']    = [ w.dot(cov).dot(w)**0.5 for w in weights ] 
random['Sharpe'] = random.RP / random.Vol     
random['Color']  = 'grey'
random['Size']   = 8

max_sharpe = random.nlargest(1,'Sharpe')
max_sharpe.Color = 'red'
max_sharpe.Size  = 50

w_ms = cov_inv.dot(rp)/cov_inv.dot(rp).sum()
ms = pd.DataFrame(index = ['Max Sharpe'])                        
ms['RP']     = [w_ms.dot(rp)]
ms['Vol']    = w_ms.dot(cov).dot(w_ms) ** 0.5      
ms['Sharpe'] = ms.RP / ms.Vol
ms['Color']  = 'aqua'
ms['Size']   = 100

w_mv = cov_inv.sum(axis=1)/cov_inv.sum()
mv = pd.DataFrame(index = ['Min Var'])                        
mv['RP']     = [w_mv.dot(rp)]
mv['Vol']    = w_mv.dot(cov).dot(w_mv) ** 0.5      
mv['Sharpe'] = mv.RP / mv.Vol
mv['Color']  = 'lime'
mv['Size']   = 100

all_portfolios = pd.concat([random, risky_assets, riskfree_asset, equal_weight, max_sharpe, ms, mv], sort=False)

graph = all_portfolios.plot.scatter('Vol','RP', s=all_portfolios.Size, color=all_portfolios.Color, xlim=(0,0.4), ylim=(0,0.16))

for t in list(rx_annual.columns):
    x = all_portfolios.loc[t,'Vol']+0.01 
    y = all_portfolios.loc[t,'RP']
    graph.text(x,y,t)
```
